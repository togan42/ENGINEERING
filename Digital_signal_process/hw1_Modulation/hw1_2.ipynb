```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
SAMPLE_RATE = 100  
DURATION = 1  

def generate_cos_wave(freq, sample_rate, duration):
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    frequencies = x * freq
    y = np.cos((2 * np.pi) * frequencies)
    return x, y
```

```python
t, x_t = generate_cos_wave(1, SAMPLE_RATE, DURATION)
plt.plot(t, x_t)
plt.show()
```

```python
fs=10
ts= 1/fs
tao= 0.05

s_t = np.zeros(100)

for i in range(100):
    
    if(i == 100):
        break
    
    if(i%10==0):
        
        s_t[i] = 1
        s_t[i+1] = 1
        s_t[i+2] = 1
        s_t[i+3] = 1
        s_t[i+4] = 1
```

```python
plt.plot(t,s_t)
```

```python
xs_t = s_t*x_t
plt.plot(t,xs_t)
```

```python
from scipy.fft import fft, fftfreq

N = SAMPLE_RATE * DURATION

yf = fft(x_t)
xf = fftfreq(N, 1 / SAMPLE_RATE)
```

```python
plt.plot(xf, np.abs(yf))
```

```python
yfs = fft(xs_t)
plt.plot(xf, np.abs(yfs))
```

```python
a_t = np.zeros(100)

for i in range(100):
    
    if(t[i%10] < 0.05):
        
        a_t[i] = 50*t[i%10] - 1.25
        
    else:
        
        a_t[i] = 3.75 - (50*t[i%10]) 
```

```python
plt.plot(t,a_t)
```

```python
x_pam = np.zeros(100)

for i in range(100):

    if(i%10==0):
        
        tut = x_t[i]

    x_pam[i] = tut 
```

```python
plt.plot(t,x_pam)
```

```python
f_t = x_pam + a_t
plt.plot(t,f_t)
```

```python
x_pdm = np.zeros(100)

for i in range(100):

    if(f_t[i]<0):

        x_pdm[i] = 0

    else:

        x_pdm[i] = 1
```

```python
plt.plot(t,x_pdm)
```

```python
yf_pdm = fft(x_pdm)
plt.plot(xf, np.abs(yf_pdm))
```
